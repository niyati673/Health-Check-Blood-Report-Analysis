```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is in a CSV file named 'blood_report.csv'
df = pd.read_csv('blood_report_with_disease.csv')

plt.scatter(df['WBC'], df['HGB'], alpha=0.5)
plt.xlabel('White Blood Cell Count')
plt.ylabel('Hemoglobin')
plt.title('WBC vs HGB Scatter Plot')
plt.show()
```

```python
plt.hist(df['RBC'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Red Blood Cell Count')
plt.ylabel('Frequency')
plt.title('Histogram of RBC Count')
plt.show()
```

```python
plt.boxplot(df['PLT'], vert=False)
plt.xlabel('Platelet Count')
plt.title('Box Plot of Platelet Count')
plt.show()
```

```python
plt.bar(df['ID'], df['MCV'], color='green')
plt.xlabel('ID')
plt.ylabel('Mean Corpuscular Volume')
plt.title('Mean Corpuscular Volume by ID')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(df['ID'], df['MCHC'], color='red')
plt.xlabel('ID')
plt.ylabel('Mean Corpuscular Hemoglobin Concentration')
plt.title('MCHC by ID')
plt.show()
```

```python
plt.pie(df['Disease'].value_counts(), labels=df['Disease'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of Diseases')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Convert 'Disease' column to numeric using one-hot encoding
# Drop the original 'Disease' column
encoded_disease = pd.get_dummies(df['Disease'], prefix='Disease', drop_first=True)
df = pd.concat([df, encoded_disease], axis=1)
df = df.drop('Disease', axis=1)

# Now calculate the correlation matrix
corr_matrix = df.select_dtypes(include=['number']).corr()  # Select only numeric columns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df[['WBC', 'HGB', 'RBC', 'PLT']])
plt.show()
```

```python
import plotly.express as px

# Choose one of the one-hot encoded columns for color
fig = px.scatter(df, x='WBC', y='HGB', color='Disease_Leukocytosis', hover_name='ID')
fig.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is in a CSV file named 'blood_report_with_disease.csv'
df = pd.read_csv('blood_report_with_disease.csv')

# Count the frequency of each disease
disease_counts = df['Disease'].value_counts()

# Create the pie chart
plt.pie(disease_counts, labels=disease_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Diseases')

plt.axis('equal')
plt.show()
```
